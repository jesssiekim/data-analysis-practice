```python
import pandas as pd

df = pd.read_csv("/spotify_2015_2025_85k.csv")
type(df)

df.head()
```

```python
#reshaping data, 가장 인기가 많은 노래 top10 을 뽑아보자

df.sort_values('popularity' , ascending=False). head(n=10)
```

```python

#생각보다 100이 많으니까 popularity가 100인 곡 수가 얼마나 많은지 알아보자
(df['popularity'] == 100).sum()
len(df[df['popularity'] == 100])
#397개가 있다고 나온다.
```

```python
#group by 함수를 이용해서 인기가 가장 많은 곡들의 장르와 국가를 보고 싶음

df[df['popularity'] == 100]['genre'].value_counts()
#힙합이 가장 많음
```
