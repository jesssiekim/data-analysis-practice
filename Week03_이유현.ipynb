```python
import pandas as pd

df = pd.read_csv("/content/spotify_2015_2025_85k.csv")
type(df)

df.head()
```

```python
#reshaping data, 가장 인기가 많은 노래 top10 을 뽑아보자

df.sort_values('popularity' , ascending=False). head(n=10)
```

```python

#생각보다 100이 많으니까 popularity가 100인 곡 수가 얼마나 많은지 알아보자
(df['popularity'] == 100).sum()
len(df[df['popularity'] == 100])
#397개가 있다고 나온다.
```

```python
#인기가 가장 많은 곡들의 장르와 국가를 보고 싶음

df[df['popularity'] == 100]['genre'].value_counts()
#힙합이 가장 많음
```

```python
df[df['popularity'] == 100]['country'].value_counts()
# 캐나다 일본, 멕시코 등등등
```

```python
#group by 함수를 이용해서 인기가 가장 많은 곡들의 장르와 국가의 조합들을 보고 싶음
df.groupby(['genre', 'country'])['popularity'].mean().sort_values(ascending=False)
```

```python
#자극적인 곡의 인기도를 살펴보자
df[df['explicit'] == 1]['popularity'].value_counts()
#보통 40점대의 인기도를 가짐
df_explicit = df[df['explicit'] == 1].sort_values(by='popularity', ascending=False)
df_explicit
```
