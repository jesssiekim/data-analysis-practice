```python
import pandas as pd

df = pd.read_csv("Electric_Vehicle_Population_Data.csv")

df.head()
```

```python
df.shape

df.columns

df.info()
```

어떤 차가 많이 등록돼 있을까? (제조사)

```python
df["Make"].value_counts().head(10)
```

모델 기준 Top 10

```python
df["Model"].value_counts().head(10)
```

제조사 + 모델 조합

```python
df.groupby(["Make", "Model"]).size().sort_values(ascending=False).head(15)
```

BEV vs PHEV 비교

```python
df["Electric Vehicle Type"].value_counts()

df["Electric Vehicle Type"].value_counts(normalize=True) * 100
```

제조사별 BEV vs PHEV 비교

```python
pd.crosstab(df["Make"], df["Electric Vehicle Type"]).head(15)
```

특정 조건 필터링 (예: TESLA만)

```python
df[df["Make"] == "TESLA"].head()

df[df["Make"] == "TESLA"]["Model"].value_counts()
```

주행거리 기준 Top 10

```python
df.sort_values(by="Electric Range", ascending=False)[
    ["Make", "Model", "Model Year", "Electric Range", "City"]
].head(10)
```

주행거리 300 이상

```python
df.sort_values(by="Electric Range", ascending=False)[
    ["Make", "Model", "Model Year", "Electric Range", "City"]
].head(10)
```

BEV만 주행거리 Top 10

```python
df[df["Electric Range"] >= 300][
    ["Make", "Model", "Electric Range"]
].sort_values(by="Electric Range", ascending=False)
```

복합 조건 필터링

예: 2020년 이후 + BEV + 200km 이상

```python
df[
    (df["Model Year"] >= 2020) &
    (df["Electric Vehicle Type"] == "Battery Electric Vehicle (BEV)") &
    (df["Electric Range"] >= 200)
][["Make", "Model", "Model Year", "Electric Range", "City"]]
```

도시별 등록 대수

```python
df["City"].value_counts().head(15)
```

특정 도시 분석 (Seattle)

```python
df[df["City"] == "Seattle"].head()

df[df["City"] == "Seattle"].shape[0]
```

도시별 + 차종 비교

```python
pd.crosstab(df["City"], df["Electric Vehicle Type"])

city = pd.crosstab(df["City"], df["Electric Vehicle Type"])
city["Total"] = city.sum(axis=1)

city.sort_values(by="Total", ascending=False).head(15)
```

특정 도시 + 제조사

```python
df[df["City"] == "Bellevue"]["Make"].value_counts().head(10)
```

복합 필터 (실전 느낌)

예: Seattle + TESLA + BEV

```python
df[
    (df["City"] == "Seattle") &
    (df["Make"] == "TESLA") &
    (df["Electric Vehicle Type"] == "Battery Electric Vehicle (BEV)")
][["Make", "Model", "Electric Range", "City"]].head(20)
```

제조사별 평균 주행거리

```python
df.groupby("Make")["Electric Range"].mean().sort_values(ascending=False).head(15)
```

모델별 평균 주행거리

```python
df.groupby(["Make", "Model"])["Electric Range"].mean().sort_values(ascending=False).head(15)
```
