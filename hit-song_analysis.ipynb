```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("spotify_2015_2025_85k.csv")
df.head()
```

```python
print(df.shape)
print(df.columns)
df.info()
```

```python
df = df.dropna(subset=["genre", "popularity", "stream_count"])
df.info()
```

Most Common Genres

```python
df = df.dropna(subset=["genre", "popularity", "stream_count"])
df.info()
```

Average Popularity By Genre

```python
genre_popularity = (
    df.groupby("genre")["popularity"]
      .mean()
      .sort_values(ascending=False)
)

genre_popularity.head(10)
```

Plot Top 10 Genres By Average Popularity

```python
genre_popularity.head(10).plot.bar(title="Top 10 Genres by Average Popularity")
plt.xlabel("Genre")
plt.ylabel("Average Popularity")
plt.tight_layout()
plt.show()
```

Average Stream Count By Genre

```python
genre_streams = (
    df.groupby("genre")["stream_count"]
      .mean()
      .sort_values(ascending=False)
)

genre_streams.head(10)
```

Plot Top 10 Genres By Average Streams

```python
genre_streams.head(10).plot.bar(title="Top 10 Genres by Average Stream Count")
plt.xlabel("Genre")
plt.ylabel("Average Stream Count")
plt.tight_layout()
plt.show()
```

Compare Explicit vs Non-explicit Songs

```python
explicit_analysis = (
    df.groupby("explicit")[["popularity", "stream_count"]]
      .mean()
      .sort_values(by="popularity", ascending=False)
)

explicit_analysis
```

Plot Explicit vs Non-explicit Popularity

```python
explicit_analysis["popularity"].plot.bar(title="Popularity: Explicit vs Non-Explicit")
plt.xlabel("Explicit")
plt.ylabel("Average Popularity")
plt.tight_layout()
plt.show()
```

Find Top Songs By Popularity

```python
top_popular_songs = (
    df[["track_name", "artist_name", "genre", "popularity", "stream_count"]]
      .sort_values("popularity", ascending=False)
      .head(10)
)

top_popular_songs
```

Find Top Songs By Streams

```python
top_streamed_songs = (
    df[["track_name", "artist_name", "genre", "popularity", "stream_count"]]
      .sort_values("stream_count", ascending=False)
      .head(10)
)

top_streamed_songs
```

Add “hit” Column

```python
popularity_cut = df["popularity"].mean()

df = df.assign(hit = df["popularity"] >= popularity_cut)

df[["track_name", "popularity", "hit"]].head()
```

Compare Hit vs Non-hit Songs

```python
hit_comparison = (
    df.groupby("hit")[["danceability", "energy", "tempo", "popularity", "stream_count"]]
      .mean()
)

hit_comparison
```

Filter Only Highly Popular Songs

```python
very_popular = df.query("popularity >= 80")
very_popular[["track_name", "artist_name", "genre", "popularity"]].head(20)
```

Short Summary

```python
print("Top genres by popularity:")
print(genre_popularity.head(5))

print("\nTop genres by streams:")
print(genre_streams.head(5))

print("\nExplicit vs non-explicit:")
print(explicit_analysis)
```
