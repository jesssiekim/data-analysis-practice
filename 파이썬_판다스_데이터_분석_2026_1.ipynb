### 2. 스트리밍 횟수가 가장 높은 상위 10곡 찾기

`stream_count` 열을 기준으로 상위 10개의 곡을 찾아봅니다.

```python
top_10_streamed_songs = dfv2.nlargest(10, 'stream_count')[['track_name', 'artist_name', 'stream_count']]
display(top_10_streamed_songs)
```

### 3. 스트리밍 횟수가 가장 높은 상위 10 아티스트 찾기

`artist_name`으로 그룹화하고 `stream_count`의 합계를 기준으로 상위 10명의 아티스트를 찾아봅니다.

```python
top_10_streamed_artists = dfv2.groupby('artist_name')['stream_count'].sum().nlargest(10).reset_index()
display(top_10_streamed_artists)
```

### 4. 스트리밍 횟수가 가장 높은 상위 10 장르 찾기

`genre`로 그룹화하고 `stream_count`의 합계를 기준으로 상위 10개의 장르를 찾아봅니다.

```python
top_10_streamed_genres = dfv2.groupby('genre')['stream_count'].sum().nlargest(10).reset_index()
display(top_10_streamed_genres)
```

### 5. 장르별 평균 인기(popularity) 확인

각 장르의 평균적인 `popularity`를 계산하여 인기 있는 장르를 파악합니다.

```python
avg_popularity_by_genre = dfv2.groupby('genre')['popularity'].mean().sort_values(ascending=False).reset_index()
display(avg_popularity_by_genre.head(10))
```

### 6. 연도별 출시 곡 수 확인

`release_year`를 기준으로 그룹화하여 각 연도에 출시된 곡의 수를 확인합니다.

```python
songs_per_year = dfv2.groupby('release_year').size().sort_values(ascending=False).reset_index(name='song_count')
display(songs_per_year.head(10))
```

### 7. 국가별 곡 수 확인

`country`를 기준으로 그룹화하여 각 국가의 곡 수를 확인합니다.

```python
songs_per_country = dfv2.groupby('country').size().sort_values(ascending=False).reset_index(name='song_count')
display(songs_per_country.head(10))
```

### 8. 장르별 평균 곡 길이(duration_ms) 확인

`genre`를 기준으로 그룹화하여 각 장르의 평균 곡 길이를 계산합니다.

```python
avg_duration_by_genre = dfv2.groupby('genre')['duration_ms'].mean().sort_values(ascending=False).reset_index()
display(avg_duration_by_genre.head(10))
```
